```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
df = pd.read_csv('Heatmap_Data.csv')
df.head()
```

```python
#extract all numeric values
numeric_df = df.iloc[:, 0:11]
numeric_df.head()
```

```python
corr = numeric_df.corr()
corr
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(corr, square=True, cmap= 'hsv', annot= True)
plt.show()
```

```python
# Plot the heatmap
fig, ax = plt.subplots(figsize=(10, 10))  # Keep the size of the figure
sns.heatmap(corr, square=True, cmap='hsv', annot=True, ax=ax, cbar=False)

# Create a divider for the existing axes instance
divider = make_axes_locatable(ax)

# Append a new axis for the color bar on the right with the same height as the heatmap
cax = divider.append_axes("right", size="5%", pad=0.5)

# Create the color bar in the newly created axis
cbar = plt.colorbar(ax.collections[0], cax=cax)
cbar.ax.tick_params(labelsize=12)

# Show the plot
plt.show()
```
